```python
import nltk
import string
from collections import Counter

def get_tokens():
    with open(r"C:\Users\Vasala Vignesh\Documents\sem5\NLP\Hyderabadi Biryani.txt",
              "r", encoding="utf-8") as food:
        text = food.read()
        lowers = text.lower()
        no_punctuation = lowers.translate(str.maketrans("", "", string.punctuation))
        tokens = nltk.word_tokenize(no_punctuation)
        return tokens

tokens = get_tokens()
count = Counter(tokens)
print(count.most_common(10))
```

```python
from nltk.corpus import stopwords
from collections import Counter

tokens = get_tokens()
filtered = [w for w in tokens if w not in stopwords.words('english')]

count = Counter(filtered)
print(count.most_common(100))
```

```python
from nltk.stem.porter import PorterStemmer

def stem_tokens(tokens, stemmer):
    return [stemmer.stem(word) for word in tokens]

stemmer = PorterStemmer()

stemmed = stem_tokens(filtered, stemmer)

count = Counter(stemmed)
print(count.most_common(100))
```

```python
import nltk
import string
import os

from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.stem.porter import PorterStemmer

path = r"C:\Users\Vasala Vignesh\Documents\sem5\NLP"

token_dict = {}
stemmer = PorterStemmer()

def stem_tokens(tokens, stemmer):
    return [stemmer.stem(item) for item in tokens]

def tokenize(text):
    tokens = nltk.word_tokenize(text)
    return stem_tokens(tokens, stemmer)

for subdir, dirs, files in os.walk(path):
    for file in files:

        if file == "Hyderabadi Biryani.txt":

            file_path = os.path.join(subdir, file)

            with open(file_path, "r", encoding="utf-8") as food:
                text = food.read()

            lowers = text.lower()
            no_punctuation = lowers.translate(
                str.maketrans("", "", string.punctuation)
            )

            token_dict[file] = no_punctuation

tfidf = TfidfVectorizer(tokenizer=tokenize, stop_words='english')
tfs = tfidf.fit_transform(token_dict.values())

print("Files processed:", list(token_dict.keys()))

print("\nTF-IDF Matrix:")
print(tfs.toarray())

print("\nWords:")
print(tfidf.get_feature_names_out())
```

```python
sentence = "Hyderabadi biryani has delicious rice and spices"

response = tfidf.transform([sentence])

print(response)

feature_names = tfidf.get_feature_names_out()

for col in response.nonzero()[1]:
    print(feature_names[col], "-", response[0, col])
```
